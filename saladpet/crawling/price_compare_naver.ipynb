```python
import time
from selenium import webdriver
import re
import pandas as pd
import time, random
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
import pandas as pd
###

import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_excel("c:\import\8total_product3.xlsx")
names = df["product_name"].tolist()
len(names)
```

```python
result = []
#names = ['dkvndk','로얄캐닌 인도어 어덜트','퓨리나 어덜트']
for i, name in enumerate(names) :
    
    names[i].replace(" ","+")
    driver = webdriver.Chrome('./chromedriver95.exe')
    driver.get("https://search.shopping.naver.com/search/all?query={}".format(names[i]))
    try :
        가격비교개수전 = driver.find_elements_by_css_selector('span[class="subFilter_num__2x0jq"]')
        가격비교개수 = int(가격비교개수전[1].text)
        
        if 가격비교개수 >= 1 : #1)가격비교 있는경우
            
            가격비교탭 = driver.find_elements_by_css_selector('a[class="subFilter_filter__3Y-uy"]')[1]
            가격비교탭.click() #가격비교 탭 클릭
            
            time.sleep(random.randint(2, 3))
            
            ### 가격비교이름중에 검색어랑 같은거 있으면 그거클릭
            title_list = []
            for data in driver.find_elements_by_css_selector('a[class="basicList_link__1MaTN"]') :
                title_list.append(data.text)
            if name in title_list :
                for data in driver.find_elements_by_css_selector('a[class="basicList_link__1MaTN"]') :
                    if data.text.strip() == name :
                        data.click()
            else :
                가격비교중첫번째링크 = driver.find_elements_by_css_selector('a[class="basicList_link__1MaTN"]')[0]
                가격비교중첫번째링크.click() #가격비교 중 첫번째 링크 클릭 (새로운 창 열기)
                
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(2)
            
            if driver.find_elements_by_css_selector("span[class='filter_text__3m_XA']") : #1-1)가격비교 페이지에 무게기준 정보가 있다면
                pre_naver_name = driver.find_element_by_css_selector('div[class="top_summary_title__15yAr"]')
                naver_name = pre_naver_name.find_element_by_css_selector('h2').text
                pre_pcode = driver.find_element_by_css_selector('div[class="naver-splugin"]')
                pcode = pre_pcode.get_attribute('data-i')
                result.append([df['product_id'][i],name,pcode,naver_name])
            driver.quit()
        else :
            result.append([df['product_id'][i],name," "," "])
            driver.quit()
    except :
        result.append([df['product_id'][i],name," "," "])
        driver.quit()
```
