```python
# 선형성(빨간 실선이 점선에서 크게 벗어나지 않는다면 예측값에 따른 잔차가 일정함으로 선형성 만족)---------------------------------
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["figure.figsize"] = (30,15)

summary = smf.ols(formula = 'crud_fiber ~ crud_protein', data = data_phos3).fit()
#예측값
model = LinearRegression()
x = data_phos3[['crud_protein']]
y = data_phos3[['crud_fiber']]
model.fit(x,y)
fitted = model.predict(x)
#잔차(실데이터 - 예측값)
residual = data_phos3[['crud_fiber']]- fitted

sns.regplot(x=fitted, y=residual, lowess = True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0,0], '--', color='gray')
plt.grid()
plt.show()


summary.summary()


```

```python
#정규성 (샤피로 검정의 pvalue가 0.05보다크면 잔차의 정규성 충족)--------------------------------------------
sp.stats.shapiro(residual)
```

```python
#잔차의 등분산성 (빨간색 실선<잔차>이 수평에 가까울수록 등분산성이 있는것이다)--------------------------------
sns.regplot(x = fitted, y=np.sqrt(np.abs(residual)), lowess = True, line_kws={'color': 'red'})
plt.grid()
plt.show()
```

```python
#잔차의 독립성(durbin watson 계수가 1.5~2.5사이일떄 : 다만 이건 다중회귀분석에서 이용하는거임)------------------
```
