```python
#이름에서 무게정보 추출하는 함수

name1 = '네츄럴코어 전연령 양고기 ECO1 7kg 강아지사료 S(8mm)'

def get_weight_from_name(namename) :
    
    if re.findall('\d+'+'kg',namename, re.I) :
        if re.findall('\d+'+'.'+'\d+'+'kg',namename, re.I) :
            name1_final = re.findall('\d+'+'.'+'\d+'+'kg',namename, re.I)
            name1_final[0] = name1_final[0].lower()
        else :
            name1_final = re.findall('\d+'+'kg',namename, re.I)
            name1_final[0] =name1_final[0].lower()

    elif re.findall('\d+'+'g',namename, re.I) :
        if re.findall('\d+'+'.'+'\d+'+'g',namename, re.I) :
            name1_final = re.findall('\d+'+'.'+'\d+'+'g',namename, re.I)
            name1_final[0] =name1_final[0].lower()
        else :
            name1_final = re.findall('\d+'+'g',namename, re.I)
            name1_final[0] =name1_final[0].lower()

    return name1_final[0]

get_weight_from_name(name1)
```
