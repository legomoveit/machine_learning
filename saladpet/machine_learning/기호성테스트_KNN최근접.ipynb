```python
## 정규화
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
#scaler = StandardScaler
variables = all_product[['p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' ,'p_pig', 'p_sheep' , 'p_duck' ,'p_salmon' , 'p_tuna'  ,'p_insect', 'o_chicken' , 'o_fish']]
scaled_variables = scaler.fit_transform(variables)
scaled_variables =  scaled_variables.tolist()



product_id_list = all_product['product_id'].tolist()

result = []

for i, data in enumerate(scaled_variables) :
    pre_result = []
    pre_result.append(product_id_list[i])
    for number in range(12) :
        pre_result.append(data[number])
    result.append(pre_result)

sample = []
for data in result :
    if data[0] in [	9	,
	285	,
	418	,
	565	,
	1468	,
	9311	,
	9627	,
	15247	,
	15347	,
	15590	] :
        sample.append(data)
    





column = ['product_id', 'p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' ,'p_pig', 'p_sheep', 'p_duck' ,'p_salmon' , 'p_tuna'  ,'p_insect', 'o_chicken' , 'o_fish']
scaled_df = pd.DataFrame.from_records( result , columns = column )

#scaled_df.to_excel('scaled_df.xlsx')
```

```python
# 0. 데이터 불러오기

test_result = pd.read_excel("c:\import\gihosung\s_0124_testresult.xlsx",sheet_name = 'result')
test_result2 = pd.read_excel("c:\import\gihosung\s_0124_testresult.xlsx",sheet_name = 'result2') #빅독제외
scaled_result = pd.read_excel("c:\import\gihosung\s_0124_testresult.xlsx",sheet_name = 'scaled_result')

df = scaled_result

x = df[['p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' ,'p_pig', 'p_sheep', 'p_duck' ,'p_salmon' , 'p_tuna'  ,'p_insect', 'o_chicken' , 'o_fish']]
y = df[['제리']]
feature_list = ['p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' ,'p_pig', 'p_sheep' ,  'p_salmon' , 'p_tuna'  ,'p_insect', 'o_chicken' , 'o_fish' ]
######
```

```python
# 2. 모델 학습 및 결과도출

clf = KNeighborsRegressor(n_neighbors = 1, weights = "distance")
clf.fit(x, y)

pred = clf.predict(scaled_df[[ 'p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' ,'p_pig', 'p_sheep', 'p_duck' ,'p_salmon' , 'p_tuna'  ,'p_insect', 'o_chicken' , 'o_fish']])
pred
#######
y_pred = list(pred)

df = pd.read_excel("제리.xlsx")
df['knn_score'] = 0

for index,score in enumerate(y_pred) :
    #print(index, score)
    df['knn_score'][index] = score

df
df.to_excel("제리_final_1.xlsx")
```
