```python
df = cat_dongdong

x = df[['p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' , 'p_sheep' ,  'p_salmon' , 'p_tuna'  , 'o_chicken' , 'o_fish' , 'flavor']]
y = df[['little_score']]
feature_list = ['p_cow' , 'p_chicken' , 'p_fish' , 'p_turkey' , 'p_sheep' ,  'p_salmon' , 'p_tuna'  , 'o_chicken' , 'o_fish' , 'flavor']
```

```python

corealtion = []
for feature in feature_list :
    a = df[feature].tolist()
    b = df['little_score'].tolist()
    number = stats.pearsonr(a,b)[0]
    zel = abs(number)
    corealtion.append([feature, number,zel])
```

```python
column = ['feature','corr','sort']
corr = pd.DataFrame.from_records( corealtion , columns = column )
corr.sort_values(by=['sort'],axis = 0,ascending = False,inplace = True)
ascending = corr['feature'].tolist()
```

```python
result = []
feature_set = []
for i in range(len(ascending)) :
    pre_feature_set = []
    for ii in range(0,i) :
        pre_feature_set.append(ascending[ii])
        
    feature_set.append(pre_feature_set)

for i in range(len(feature_set)) :
    
    if i == 0 :
        pass
    
    elif i == 1 :
        model = smf.ols(formula = 'little_score ~' + feature_set[i][0],data = df).fit()
        result.append([feature_set[i], model.rsquared, model.rsquared_adj, model.aic, model.bic])
        
    else :
        feature_set_plus = '+'.join(feature_set[i])
        model = smf.ols(formula = 'little_score ~' + feature_set_plus,data = df).fit()
        result.append([feature_set[i], model.rsquared, model.rsquared_adj,model.aic, model.bic])
        
    #model.rsqaured_adj, 
```
